# Human vs random

```python
import random
import pickle
name=input('please enter your name')
try:
    with open(f'{name}_vs_random.pickle','rb') as e:
        step=pickle.load(e)
except:
    step=[]
for _ in range(30):
    winner=2
    turn=0
    map=[0 for n in range(9)]
    cir=[" ","O","X"]
    num=[n for n in range(1,10)]
    while winner!=0 and winner!=1 and map.count(0):
        graph=f'''
{cir[map[0]]}|{cir[map[1]]}|{cir[map[2]]}
-+-+-
{cir[map[3]]}|{cir[map[4]]}|{cir[map[5]]}
-+-+-
{cir[map[6]]}|{cir[map[7]]}|{cir[map[8]]}
'''
        print(graph)
        if turn%2==0:
            place=int(input('please pick a spot 1~9'))
        else:
            place=random.choice(num)
        try:
            num.remove(place)
        except:
            place=int(input('please pick a spot 1~9'))
        place-=1
        if turn%2==0:
            record=[n for n in map]
            step.append([record,place+1])
        map[place]=turn%2+1
        if map[place%3]==map[place%3+6] and map[place%3+3]==map[place%3]:
            winner=turn%2
        elif map[place//3*3+1]==map[place//3*3+2] and map[place//3*3]==map[place//3*3+1]:
            winner=turn%2
        elif map[4]==map[0] and map[0]==map[8] and map[0]==map[place]:
            winner=turn%2
        elif map[4]==map[2] and map[2]==map[6] and map[2]==map[place]:
            winner=turn%2
        if winner!=2:
            print(f'player {turn%2+1} win!!!')
        elif map.count(0)==0:
            print("even")

            
        turn+=1
    with open(f'{name}_vs_random.pickle','wb') as e:

        pickle.dump(step,e)
        print('saved')
```

# Player vs Player

```python
import random
import pickle
name=input('please enter your name')
try:
    with open(f'{name}_vs_human_f.pickle','rb') as e:
        fstep=pickle.load(e)
    with open(f'{name}_vs_human_l.pickle','rb') as e:
        lstep=pickle.load(e)       
except:
    fstep=[]
    lstep=[]
for _ in range(20):
    winner=2
    turn=0
    map=[0 for n in range(9)]
    cir=[" ","O","X"]
    num=[n for n in range(1,10)]
    while winner!=0 and winner!=1 and map.count(0):
        graph=f'''
{cir[map[0]]}|{cir[map[1]]}|{cir[map[2]]}
-+-+-
{cir[map[3]]}|{cir[map[4]]}|{cir[map[5]]}
-+-+-
{cir[map[6]]}|{cir[map[7]]}|{cir[map[8]]}
'''
        print(graph)
        if turn%2==0:
            place=int(input('please pick a spot 1~9 player1'))
        else:
            place=int(input('please pick a spot 1~9 player2'))
        try:
            num.remove(place)
        except:
            place=int(input('please pick another spot 1~9'))
        place-=1
        s=[n for n in map]
        if turn%2==0:
            fstep.append([s,place+1])
        else:
            lstep.append([s,place+1])
        map[place]=turn%2+1
        if map[place%3]==map[place%3+6] and map[place%3+3]==map[place%3]:
            winner=turn%2
        elif map[place//3*3+1]==map[place//3*3+2] and map[place//3*3]==map[place//3*3+1]:
            winner=turn%2
        elif map[4]==map[0] and map[0]==map[8] and map[0]==map[place]:
            winner=turn%2
        elif map[4]==map[2] and map[2]==map[6] and map[2]==map[place]:
            winner=turn%2
        if winner!=2:
            print(f'player {turn%2+1} win!!!')
        elif map.count(0)==0:
            print("even")
        turn+=1
with open(f'{name}_vs_human_first.pickle','wb') as e:
    pickle.dump(fstep,e)
    print('saved')
with open(f'{name}_vs_human_last.pickle','wb') as e:
    pickle.dump(lstep,e)
    print('saved')
```

# AI testing

```python
from sklearn.ensemble import RandomForestClassifier
import sklearn
import pickle
name=input('please enter your name')
with open(f'{name}_vs_random.pickle','rb') as rand:
    win=pickle.load(rand)
with open(f'{name}_vs_human_first.pickle','rb') as e:
    win.extend(pickle.load(e))
x=[n[0] for n in win]
y=[n[1] for n in win]
clf=RandomForestClassifier(n_estimators=30)
clf.fit(x,y)
cir=[" ","O","X"]
with open(f'{name}.pickle','wb') as e:
    pickle.dump(clf,e)
for _ in range(5):
    winner=2
    turn=0
    map=[0 for n in range(9)]
    num=[n for n in range(1,10)]
    while winner==2 and map.count(0):
        graph=f'''
{cir[map[0]]}|{cir[map[1]]}|{cir[map[2]]}
-+-+-
{cir[map[3]]}|{cir[map[4]]}|{cir[map[5]]}
-+-+-
{cir[map[6]]}|{cir[map[7]]}|{cir[map[8]]}'''
        print(graph)
        if turn%2==0:
            place=clf.predict([map])[0]
        else:
            place=int(input('please enter 1~9'))
        place-=1
        if map[place]==0:
            num.remove(place+1)
            map[place]=turn%2+1
            
            if map[place%3]==map[place%3+6] and map[place%3+3]==map[place%3]:
                winner=turn%2
            elif map[place//3*3+1]==map[place//3*3+2] and map[place//3*3]==map[place//3*3+1]:
                winner=turn%2
            elif map[4]==map[0] and map[0]==map[8] and map[0]==map[place]:
                winner=turn%2
            elif map[4]==map[2] and map[2]==map[6] and map[2]==map[place]:
                winner=turn%2
            if winner!=2:
                graph=f'''
{cir[map[0]]}|{cir[map[1]]}|{cir[map[2]]}
-+-+-
{cir[map[3]]}|{cir[map[4]]}|{cir[map[5]]}
-+-+-
{cir[map[6]]}|{cir[map[7]]}|{cir[map[8]]}'''
                print(graph)
                print(f'winner is player {turn%2+1}')
        
                
        else:
            winner=0
            print('error')
        turn+=11
```
